### Summarize the metrics of all experiments

```python
target_results = [ target_dir for target_dir in os.listdir(os.path.join(base_dir, 'result')) if target_dir.startswith("yad_rest") ] 
dfs = []
for target_result in target_results:
    result_dir = os.path.join(base_dir, "result", target_result)
    print(target_result)
    #target_samples = torch.load(os.path.join(result_dir, "samples.pkl"))
    try:
        args = pd.read_csv(os.path.join(result_dir, "argv.csv"), header=None)

        metric_df = pd.read_csv(os.path.join(result_dir, "metric.csv"))
        metric_len = metric_df.shape[0]
        metric_mean = metric_df[['accuracy','recall', 'precision','roc_auc']].loc[metric_len-2].map('{:.3f}'.format)
        metric_std = metric_df[['accuracy','recall', 'precision','roc_auc']].loc[metric_len-1].map('{:.3f}'.format)
        metric_summary = metric_mean + u'\xb1' + metric_std
        dfs.append(pd.DataFrame({target_result: metric_summary}).T)
    except:
        print(f"Results for {target_result} are not found.")
df = pd.concat(dfs, axis=0)        
df = df.sort_index()
#df.to_csv(os.path.join(base_dir, 'result', 'metric_summary.csv'), encoding='utf-8')
df
```
